```python
import pandas as pd

# FIRST DATAFRAME
df1 = pd.DataFrame({
    "rollno": [1, 2, 3, 4],
    "name": ["Aman", "Neha", "Ravi", "Sara"],
    "subject": ["Math", "Science", "English", "CS"],
    "marks": [85, 90, 78, 88]
})

# SECOND DATAFRAME
df2 = pd.DataFrame({
    "rollno": [1, 2, 3, 4],
    "department": ["CSE", "ECE", "ME", "IT"],
    "annual": [300000, 350000, 280000, 400000],
    "father": ["Ram", "Shyam", "Mohan", "Amit"]
})

#  SET INDEX
df1 = df1.set_index("rollno")
df2 = df2.set_index("rollno")

#  MERGE USING INDEX
result = pd.merge(df1, df2, left_index=True, right_index=True)
print(result)
```
