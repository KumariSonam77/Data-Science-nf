```python
#merge

import pandas as pd
Df1 = pd.DataFrame({"A":[1,2,3,4],"B":["A","B","C","D"]})
Df2 = pd.DataFrame({"A":[1,2,3,4],"D":["a","b","c","d"]})
pd.merge(Df1,Df2)
```

```python
#merge follow on value

Df1 = pd.DataFrame({"A":[1,2,3,4],"B":["A","B","C","D"]})
Df2 = pd.DataFrame({"A":[1,2,3,4],"B":["a","b","c","d"]})
pd.merge(Df1,Df2,on="B")
```

```python
#inner merge
Df1 = pd.DataFrame({"A":[1,2,3,4],"B":["A","B","C","D"]})
Df2 = pd.DataFrame({"A":[1,2,3,4],"C":["a","b","c","d"]})
pd.merge(Df1,Df2,on="A",how = 'inner',indicator=True)
```

```python
#outer merge(union)
Df1 = pd.DataFrame({"A":[1,2,3,4],"B":["A","B","C","D"]})
Df2 = pd.DataFrame({"A":[1,2,3,4],"C":["a","b","c","d"]})
pd.merge(Df1,Df2,on="A",how = 'outer',indicator=True)
```

```python
#left merge ()
Df1 = pd.DataFrame({"A":[1,2,3,4],"B":["A","B","C","D"]})
Df2 = pd.DataFrame({"A":[1,5,7,4],"C":["a","b","c","d"]})
pd.merge(Df1,Df2,on="A",how = 'left',indicator=True)
```

```python
#right merge
Df1 = pd.DataFrame({"A":[1,2,3,4],"B":["A","B","C","D"]})
Df2 = pd.DataFrame({"A":[1,5,7,4],"C":["a","b","c","d"]})
pd.merge(Df1,Df2,on="A",how = 'right',indicator=True)
```

```python
Df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":["A","B","C","D","E"]})
Df2 = pd.DataFrame({"A":[1,5,7,4,5],"B":["a","b","c","d","e"]})
pd.merge(Df1,Df2,left_index=True,right_index=True)
```

```python
#remove x,y add name use suffix
Df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":["A","B","C","D","E"]})
Df2 = pd.DataFrame({"A":[1,5,7,4,5],"B":["a","b","c","d","e"]})
pd.merge(Df1,Df2,left_index=True,right_index=True,suffixes=('Id','name'))
```

```python
import pandas as pd

Df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":["A","B","C","D","E"]})
Df2 = pd.DataFrame({"C":[1,5,7,4,5],"D":["a","b","c","d","e"]})
Df2.join(Df1)
```

```python
#index in join

Df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":["A","B","C","D","E"]},index=["@","#","a",2,"^"])
Df2 = pd.DataFrame({"C":[1,5,7,4,5],"D":["a","b","c","d","e"]},index=["@","#","a",2,"^"])
Df2.join(Df1)
```

```python
#change index value
Df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":["A","B","C","D","E"]},index=["@","#","a",2,"^"])
Df2 = pd.DataFrame({"C":[1,5,7,4,5],"D":["a","b","c","d","e"]},index=["^",2,"b",3,"*"])
Df2.join(Df1)
```

```python
#inner join
Df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":["A","B","C","D","E"]})
Df2 = pd.DataFrame({"C":["A","B"], "D":["E","F"]})
Df1.join(Df2,how='inner')
```

```python
#outer join
Df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":["A","B","C","D","E"]})
Df2 = pd.DataFrame({"C":["A","B"], "D":["E","F"]})
Df1.join(Df2,how='outer')
```

```python
#left join
Df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":["A","B","C","D","E"]})
Df2 = pd.DataFrame({"C":["A","B"], "D":["E","F"]})
Df1.join(Df2,how='left')
```

```python
#right join
Df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":["A","B","C","D","E"]})
Df2 = pd.DataFrame({"C":["A","B"], "D":["E","F"]})
Df1.join(Df2,how='right')
```

```python
#suffixes in join (left and right)
Df1 = pd.DataFrame({"A":[1,2,3,4,5],"B":["A","B","C","D","E"]})
Df2 = pd.DataFrame({"A":[1,5,7,4,5],"B":["a","b","c","d","e"]})
Df1.join(Df2,lsuffix=True,rsuffix=True)
```

```python
import pandas as pd

# FIRST DATAFRAME
df1 = pd.DataFrame({
    "rollno": [1, 2, 3, 4],
    "name": ["Aman", "Neha", "Ravi", "Sara"],
    "subject": ["Math", "Science", "English", "CS"],
    "marks": [85, 90, 78, 88]
})

# SECOND DATAFRAME
df2 = pd.DataFrame({
    "rollno": [1, 2, 3, 4],
    "department": ["CSE", "ECE", "ME", "IT"],
    "annual": [300000, 350000, 280000, 400000],
    "father": ["Ram", "Shyam", "Mohan", "Amit"]
})

#  SET INDEX
df1 = df1.set_index("rollno")
df2 = df2.set_index("rollno")

#  MERGE USING INDEX
result = pd.merge(df1, df2, left_index=True, right_index=True)
print(result)
```
