```python
## Step 1: Load and Inspect Dataset
#First, I loaded the dataset into Pandas. Then I checked the first few rows to understand what data is available in the dataset.

import pandas as pd
df = pd.read_csv("Messy_Employment_India_Dataset.csv")
df.head()
```

```python
## Step 2: Understand Dataset Structure and Data Types
#This step helps us understand the number of records, available columns, data types, and missing values.

df.info()
```

```python
df.shape
```

```python
df.columns
```

```python
## Step 3: Identify and Handle Missing Values
#Missing values reduce data quality. Categorical columns are filled using the mode, while numerical columns are filled using the median.
df.isnull().sum()
```

```python
## Step 4: Remove Duplicate Records
#Duplicate records can lead to inaccurate analysis. These records are identified and removed.

print("Duplicate Rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Duplicate Rows After Cleaning:", df.duplicated().sum())
```

```python
## Step 5: Correct Inconsistent Formatting
#Text values often contain extra spaces and inconsistent capitalization. These issues are corrected.

for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.strip()
```

```python
## Step 6: Fix Data Type Issues
#The Date Recorded column should be stored as a datetime data type for proper analysis.

df['Date Recorded'] = pd.to_datetime(
    df['Date Recorded'],
    errors='coerce'
)
df.info()
```

```python
## Step 7: Detect and Handle Outliers
#Outliers in salary values are identified using the IQR method and removed.

Q1 = df['Monthly Salary (INR)'].quantile(0.25)
Q3 = df['Monthly Salary (INR)'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[
    (df['Monthly Salary (INR)'] >= lower) &
    (df['Monthly Salary (INR)'] <= upper)
]
```

```python
## Step 8: Rename and Standardize Column Names
#Column names are converted into lowercase format and spaces are replaced with underscores.
df.columns = (
    df.columns
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("(", "", regex=False)
      .str.replace(")", "", regex=False)
)
df.columns
```

```python
## Step 9: Validate and Verify Cleaned Data
#The cleaned dataset is checked again to verify that all cleaning operations were successful.

df.info()
```

```python
## Step 10: Export Final Cleaned Dataset
#The final cleaned dataset is exported as a CSV file for future use.

df.to_csv(
    "Cleaned_Employment_India_Dataset.csv",
    index=False
)
print("Dataset exported successfully!")
```
