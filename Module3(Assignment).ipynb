What is Matplotlib?

Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations. It helps users represent data graphically through charts, plots, histograms, pie charts, scatter plot.

Purpose of Matplotlib

*  Data visualization and analysis
*  Creating publication-quality graphs
*  Exploring trends and patterns in datasets
*  Comparing datasets visually
*  Supporting scientific and business reporting

10 Commonly Used Modules of Matplotlib Module Functionality

1.pyplot = Used to create graphs and charts easily.

2.figure = Used to create and manage the main figure  where plots are drawn.

3.axes = Controls the plotting area, axis labels, titles, and limits.

4.colors = Used to define and customize colors in plots.

5.show = function is used to display the graph or plot on the screen.

6.lines = Used to create and customize line objects.

7.ticks = Ticks are the marks and labels shown on the X-axis and Y-axis.

8.annotation = Annotation is used to add notes, labels, or comments to specific points on a graph.

9.grid = A grid adds horizontal and vertical reference lines to make the chart easier to read.

10.title = A title describes what the chart represents.







#2.Create and demonstrate both a Double Bar Chart and a Grouped Bar Chart using Matplotlib.

1. Double Bar Chart

A Double Bar Chart compares two datasets for the same categories.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Exam1": [50, 60, 70, 80],
    "Exam2": [55, 65, 75, 85]
}, index=["Python", "JavaScript", "React.js", "TypeScript"])

df.plot(kind="bar", figsize=(7,5))

plt.title("Double Bar Chart")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.show()
```

2. Grouped Bar Chart
A Grouped Bar Chart compares three or more datasets for the same categories.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Exam1": [50, 60, 70, 80],
    "Exam2": [55, 65, 75, 85],
    "Exam3": [60, 70, 80, 90]
}, index=["Python", "JavaScript", "React.js", "TypeScript"])

# Grouped Bar Chart
df.plot(kind="bar", figsize=(8,5))

plt.title("Grouped Bar Chart")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.legend(title="Exams")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.show()
```

```python
#IMPORT LIBERARIES
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("penguins.csv").head(20)
df.head(20)
```

2. Violin Plot-
A Violin Plot combines a Box Plot and a Density Plot. It shows the distribution of data and how frequently values occur.

```python
sns.violinplot(data= df, x="species", y="body_mass_g", hue= "sex")
plt.title("Violin Plot")
plt.show()
```

3. Box Plot
A Box Plot displays the distribution of data using five statistical measures:

```python
sns.boxplot(data= df, x="species", y="body_mass_g", hue= "sex")
sns.despine()
plt.title("Box Plot")
plt.show()
```

4. KDE (Kernel Density Estimation) Plot

A KDE Plot is a smooth curve that represents the probability density of a dataset. It is similar to a histogram but provides a smoother view of the distribution.

```python
sns.kdeplot(data= df, x="body_mass_g", hue= "sex",fill =True)
sns.despine()
```
